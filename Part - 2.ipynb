### Get dataframe of the postal code of each neighborhood

```python
df=pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")[0]
df_new=df[df['Borough'] != 'Not assigned']
df_new.reset_index(drop=True, inplace=True)
df_new.head(12)
```

### Transform the Geospatial Data into a pandas dataframe

```python
dg=pd.read_csv("https://cocl.us/Geospatial_data")
dg.head()
```

### Merging both the dataframes

```python
neighborhoods = pd.merge(df_new, dg, on='Postal Code')
neighborhoods.head(12)
```
